**Опис завдання:**
Створити клас `ConsonantCounter`, який:
* Приймає в конструкторі рядок.
* Має метод `count()`, який підраховує кількість приголосних букв (всі латинські букви крім $a, e, i, o, u$).
* Має property `consonant_count`, яка повертає підраховану кількість.

**User Prompt:**
Твоя задача виконати 'Завдання 26' дотримуючись наступних вимог:
1. Реалізувати метод `count()` для підрахунку приголосних латинських літер (ігноруючи регістр).
2. Створити property `consonant_count` для виведення результату, використовуй українську мову в функції `print`.
3. Створити метод `__str__`, який повертає рядок "Виконав: Богдан, Група: КН-33".
4. У циклі створити 3 об'єкти класу з різними вхідними рядками.
5. Дотримуйся стандартів PEP 8.

```python
class ConsonantCounter:
    def __init__(self, text: str):
        self.text = text
        self._count = None

    def count(self) -> int:
        """Підраховує кількість приголосних латинських літер."""
        vowels = "aeiou"
        consonants_count = 0
        
        for char in self.text.lower():
            # Перевірка: чи символ є латинською літерою і чи він не голосний
            if char.isalpha() and char.isascii() and char not in vowels:
                consonants_count += 1
        
        self._count = consonants_count
        return self._count

    @property
    def consonant_count(self) -> int:
        """Property для отримання та виведення результату."""
        result = self.count()
        print(f"Кількість приголосних букв у рядку: {result}")
        return result

    def __str__(self) -> str:
        return "Виконав: Данило, Група: КН-33"


# Створення 3 об'єктів класу з різними вхідними даними у циклі
test_strings = [
    "Hello World",
    "Object-Oriented Programming",
    "Python 3.12"
]

for item in test_strings:
    counter = ConsonantCounter(item)
    print(counter)
    print(f"Вхідний текст: '{item}'")
    # Виклик property
    _ = counter.consonant_count
    print("-" * 40)
```
